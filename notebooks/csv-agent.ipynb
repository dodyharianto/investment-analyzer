```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns

file_path = '../amazon_balance_sheet.csv'
df = pd.read_csv(file_path)
df
```

```python
from langchain_experimental.tools import PythonAstREPLTool

# Create a tool to execute Python codes.
python_tool = PythonAstREPLTool()

# Load the DataFrame on locals["df"]
python_tool.locals["df"] = df
```

```python
python_tool.locals['df']
```

```python
def visualize_financial_data(file_path: str, metric: str, title='figure'):
    """
    Use this to create a matplotlib figure to visualize the value of a financial metric over time.
    """
    plt.figure(figsize=(8, 5))
    python_tool.locals['df'].plot(y=metric, legend=True, marker='o')
    plt.xlabel('Date')
    plt.ylabel('Value')
    plt.grid(True)
    plt.savefig(f'../outputs/visualizations/{title}.png')
    plt.show()

    return {'figure_file_name': f'{title}.png'}
```

```python
visualize_financial_data(file_path=file_path, metric='CurrentLiabilities')
```
