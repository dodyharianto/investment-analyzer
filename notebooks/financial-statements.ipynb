```python
from yahooquery import Ticker

symbols = ['fb', 'aapl', 'amzn', 'nflx', 'goog']
stock_data = Ticker(symbols)
stock_data.summary_detail
```

```python
import pandas as pd

aapl_df = pd.DataFrame(stock_data.summary_detail)
aapl_df
```

```python
balance_df = stock_data.balance_sheet().set_index('asOfDate')
balance_df.index = pd.to_datetime(balance_df.index)
balance_df
```

```python
balance_df.info()
```

```python
balance_df.columns
```

```python
balance_df['CurrentRatio'] = balance_df['CurrentAssets'] / balance_df['CurrentLiabilities']
balance_df['QuickRatio'] = balance_df['CurrentAssets'] / balance_df['Inventory'] / balance_df['CurrentLiabilities']
balance_df['DebtToEquityRatio'] = balance_df['TotalDebt'] / balance_df['StockholdersEquity']
balance_df
```

```python
from matplotlib import pyplot as plt
import seaborn as sns


plt.figure(figsize=(8, 5))
balance_df.plot(y='CurrentAssets', legend=True, marker='o', title="Line Plot with Date Index")
plt.xlabel("Date")
plt.ylabel("Value")
plt.grid(True)
plt.show()
```
